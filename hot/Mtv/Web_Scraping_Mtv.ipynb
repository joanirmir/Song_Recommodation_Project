#### Building the dataframe

```python
#initialize empty lists
artist = []
title = []
```

```python
# define the number of iterations of our for loop 
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("article.promo-type-a > div.subject"))
print(num_iter)
```

```python
len(soup.select("article.promo-type-a > div.promo-title"))
```

```python
# iterate through the result set and retrive all the data
for i in range(num_iter):
    artist.append(soup.select("article.promo-type-a > div.subject")[i].get_text().strip()) ## getting movies titles
    title.append(soup.select("article.promo-type-a > div.promo-title")[i].get_text().strip()) ## getting dir and actors names
```

```python
# each list becomes a column
mtv = pd.DataFrame({"artist":artist,
                       "title":title})

mtv.head()
```
