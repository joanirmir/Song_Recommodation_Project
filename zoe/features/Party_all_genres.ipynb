#### building Data frame of audio features

```python
my_dict = sp.audio_features(playlist["items"][0]["track"]["uri"])[0] # you can provide a list of uri's
my_dict
```

```python
import pandas as pd

list_of_songs=[]
for index in range(0,len(playlist["items"])):
    list_of_songs.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
while playlist['next']:
    playlist = sp.next(playlist)
    for index in range(0,len(playlist["items"])):
        list_of_songs.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
pd.DataFrame(list_of_songs) 
```

```python
party_all_genres = pd.DataFrame(list_of_songs) 
```

```python
party_all_genres.to_csv("party_all_genres.csv",index=False)
```
