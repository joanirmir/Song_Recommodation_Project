#### building Data frame of audio features

```python
import pandas as pd

features=[]
for index in range(0,len(playlist["items"])):
    features.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
while playlist['next']:
    playlist = sp.next(playlist)
    for index in range(0,len(playlist["items"])):
        features.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
pd.DataFrame(features) 
```

```python
party_songs = pd.DataFrame(features) 
```

#### Extracting the songs of a playlist

```python
playlist['items'][0]['track']['name']
```

```python
playlist = sp.user_playlist_tracks("spotify", "5xS3Gi0fA3Uo6RScucyct6",market="GB")

titles = []
for index in range(0,len(playlist['items'])):  
    titles.append(playlist['items'][index]['track']['name'])
while playlist['next']: 
    playlist = sp.next(playlist)  
    for index in range(0,len(playlist['items'])):
        titles.append(playlist['items'][index]['track']['name'])
```

#### Getting the artists of the playlist 

```python
playlist['items'][0]['track']['artists']
```

```python
playlist['items'][0]['track']['artists'][0]['name']
```

```python
playlist = sp.user_playlist_tracks("spotify", "5xS3Gi0fA3Uo6RScucyct6",market="GB")

artists = []
for index in range(0,len(playlist['items'])):  
    artists.append(playlist['items'][index]['track']['artists'][0]['name'])
while playlist['next']: 
    playlist = sp.next(playlist)  
    for index in range(0,len(playlist['items'])):
        artists.append(playlist['items'][index]['track']['artists'][0]['name'])
```

#### dataframe + artist column + titles column

```python
party_songs["artists"] = artists
party_songs["titles"] = titles
```
