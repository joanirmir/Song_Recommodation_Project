#### Rap Caviar

```python
#https://open.spotify.com/playlist/37i9dQZF1DX0XUsuxWHRQd
import config
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.c_id, client_secret=config.c_se))
playlist = sp.user_playlist_tracks("spotify", "37i9dQZF1DX0XUsuxWHRQd",market="GB")
print(list(playlist.keys()))
print("Total number of songs in the playlist: ",playlist["total"]) 
len(playlist["items"])
```

```python
import pandas as pd
features=[]
for index in range(0,len(playlist["items"])):
    features.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
while playlist['next']:
    playlist = sp.next(playlist)
    for index in range(0,len(playlist["items"])):
        if sp.audio_features(playlist["items"][index]["track"]["uri"])[0]is None: #for missing songs
            continue
        else:                           
            features.append(sp.audio_features(playlist["items"][index]["track"]["uri"])[0])
rap_caviar = pd.DataFrame(features)  
```

```python
playlist = sp.user_playlist_tracks("spotify", "37i9dQZF1DX0XUsuxWHRQd",market="GB")
titles = []
for index in range(0,len(playlist['items'])):  
    titles.append(playlist['items'][index]['track']['name'])
while playlist['next']: 
    playlist = sp.next(playlist)  
    for index in range(0,len(playlist['items'])):
        if sp.audio_features(playlist["items"][index]["track"]['name'])is None: #for missing songs
            continue
        else:   
            titles.append(playlist['items'][index]['track']['name'])
```

```python
len(titles)
```

```python
playlist = sp.user_playlist_tracks("spotify", "37i9dQZF1DX0XUsuxWHRQd",market="GB")

artists = []
for index in range(0,len(playlist['items'])):  
    artists.append(playlist['items'][index]['track']['artists'][0]['name'])
while playlist['next']: 
    playlist = sp.next(playlist)  
    for index in range(0,len(playlist['items'])):
        if sp.audio_features(playlist["items"][index]["track"]['artists'][0]['name'])is None: #for missing songs
            continue
        else: 
            artists.append(playlist['items'][index]['track']['artists'][0]['name'])
```
