# Use case: Official Singles Chart Top 100

```python
# 1. import libraries
import sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 2. find url and store it in a variable
url = "https://www.officialcharts.com/charts/singles-chart/"
```

#### using request package

```python
# 3. download html with a get request
response = requests.get(url)
```

```python
response.content
```

```python
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
# 4.2. check that the html code looks like it should
soup
```

<b> you can copy the CSS selector using Chrome inspector. Just select the element you want to extract  and right click and select copy selector 

```python
soup.select("div.title-artist > div.title")[0].get_text().strip()
```

```python
soup.select("div.title-artist > div.artist")[0].get_text().strip()
```

#### Building the dataframe

```python
#initialize empty lists
title = []
artist = []
```

```python
# define the number of iterations of our for loop 
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("div.title-artist > div.title"))
print(num_iter)
```

```python
len(soup.select("div.title-artist > div.artist"))
```

```python
# iterate through the result set and retrive all the data
for i in range(num_iter):
    title.append(soup.select("div.title-artist > div.title")[i].get_text().strip()) ## getting movies titles
    artist.append(soup.select("div.title-artist > div.artist")[i].get_text().strip()) ## getting dir and actors names
```

```python
title
```

```python
artist
```

```python
# each list becomes a column
official_charts = pd.DataFrame({"title":title, 
                    "artist":artist,})

official_charts.head()
```

```python
official_charts.to_csv("official_charts.csv",index=False)
```
